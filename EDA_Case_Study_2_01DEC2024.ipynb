```python
# Importing the important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Uploading the data
from google.colab import files
uploaded = files.upload()
```

```python
# Getting the data into a dataframe
df = pd.read_csv('merc.csv')
df.head()
```

```python
# Step 1 : General Information about the dataset (Data Type and Missing Values)
df.info()
```

```python
# Step 2 : Getting the statistical properties about the data
df.describe()
```

```python
# Step 3 : Getting the idea about NULL values
df.isnull().sum().any()
```

```python
# False means that there are no NULL values in the dataset
```

```python
# Step 4 : Getting an idea about the duplicates
df.duplicated().sum()
```

```python
# 0 means that there are no duplicates value in the data
```

```python
# Step 5 : Getting the number of unique values
df.nunique()
```

```python
# Step 6 : Getting the unique values
df['model_ID'].unique()
```

```python
# Step 7 : Average price per year
df.groupby('year') ['price'].mean()
```

```python
# Median mileage per year
df.groupby('year') ['mileage'].median()
```

```python
# Step 8 : Matrix for Coorelation
df.corr()
```

```python
# Heatmap
plt.figure(figsize=(9,5)) # Setting the figuresize (in inches)
sns.heatmap(df.corr(),annot=True)
```

```python
# Step 9 : Getting the proportion of values
df['fuel_ID'].value_counts()
```

```python
# Majority of the vehicles are from 1 and 5 fuel_ID.
```

```python
# Step 10 : Data Visulization
# Histogram for the car prices
plt.figure(figsize=(9,3))
sns.histplot(df['price'], bins=10, kde = True)
plt.title('Histogram for the Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Data Visulization
# Histogram for the car mileage
plt.figure(figsize=(9,3))
sns.histplot(df['mileage'], bins=10, kde = True)
plt.title('Histogram of Mileage')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()
```
